```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear():
    x = np.linspace(-10,10,100)
    y = x
    plt.plot(x,y)
    plt.xlabel("Input")
    plt.ylabel("output")
    plt.title("Linear")
    plt.grid(True)
    plt.show()
linear()
```

```python
def sigmoid():
    x = np.linspace(-10,10,100)
    y = 1/(1+np.exp(-x))
    plt.plot(x,y)
    plt.xlabel("Input")
    plt.ylabel("output")
    plt.title("Sigmoid")
    plt.grid(True)
    plt.show()
sigmoid()
```

```python
def tanh():
    x = np.linspace(-10,10,100)
    y = np.tanh(x)
    plt.plot(x,y)
    plt.xlabel("Input")
    plt.ylabel("output")
    plt.title("Tanh")
    plt.grid(True)
    plt.show()
tanh()
```

```python
def relu():
    x = np.linspace(-10,10,100)
    y = np.maximum(0,x)
    plt.plot(x,y)
    plt.xlabel("Input")
    plt.ylabel("output")
    plt.title("RelU")
    plt.grid(True)
    plt.show()
relu()
```
